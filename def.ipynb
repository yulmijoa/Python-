```python
def get_area(radius):
  if radius>0:
    return 3.14 * radius ** 2
  else :
    return 0

radius = int(input('반지름: '))
result = get_area(radius)
print(f'반지름이 {radius}인 원의 면적 = {result}')
```

```python
def get_sum(start, end):
  s = 0
  for i in range(start, end+1):
    s += i
  return s

start = int(input('시작 값: '))
end = int(input('종료 값: '))
x = get_sum(start, end)
print(f'{start}부터 {end}까지의 정수합은 {x}입니다.')
```

```python
def displat(msg, cnt):
  for i in range(cnt):
    print(msg)

msg = input('문구: ')
cnt = int(input('출력 횟수: '))

displat(msg, cnt)
```

```python
def greet(name, msg=' 별일 없으시죠?'):
  print('안녕',name + ',' + msg)    #print면 여러개 가능, return이면 마지막 것만

greet('영희', ' 건강하지?')
greet('영희')
```

```python
def power(base, expo):
  return base ** expo   #print면 여러개 가능, return이면 마지막 것만

power(2, 10)
power(10, 2)
```

```python
def star(p, q):
  return(p*q)

p = input('원하는 기호나 숫자: ')
q = int(input('원하는 반복횟수: '))

star(p, q)
```

```python
def star(p, q, r):
  for i in range(r):
    print(p * q)
    print()

star('a', 5, 10)
```

위 코드를 for 말고 while문 써서, while이랑 break 써서

```python
i = 1
def star(p, q, r):
  while i == r:
    print()
    i += 1
star('a', 5, 10)
```

```python
import ColabTurtle.Turtle as t
t.initializeTurtle()

def square(n,length):
  for i in range(n):
    t.forward(length)
    t.left(360/n)

square(4, 100)
square(5, 200)
```

```python
def weeklyPay(rate, hour):
  if hour >= 30 :
    return hour%30 * rate * 1.5 + rate * 30
  else :
    return hour*30

rate = int(input('시급을 입력하시오: '))
hour = int(input('근무 시간을 입력하시오: '))
print(f'주급은 {weeklyPay(rate,hour)}')
```

```python
today = datetime.datetime.now()

def getCurrent():
  cur = datetime.now()
  return cur

def getAfter(now, day):
  ret = now + timedelta(days = day)
  return ret

nowDate, afterDate = None, None

nowDate = getCurrent()
print('현재 날짜 ==>', nowDate)
afterDate = getAfter(nowDate,100)
print('100일 후 날짜 ==>', afterDate)
```
