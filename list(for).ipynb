```python
seoul = 9765
busan = 3441
incheon = 2954
city_pop = [seoul, busan, incheon]

daejeon = 1531
city_pop.append(daejeon)

max_pop = 0
min_pop = 1000000
pop_sum = 0
n = 0

for pop in city_pop:
  if pop > max_pop:
    max_pop = pop
  if pop < min_pop :
    min_pop = pop
  pop_sum += pop
  n += 1

print(f'최대 인구: {max_pop}')
print(f'최소 인구: {min_pop}')
print(f'평균 인구: {pop_sum/n}')
```

```python
import random

question = []
a, b = 0, 0
op = ['+', '-', '%', '*']
for i in range(100):
  a = random.randint(1, 100)
  b = random.randint(1, 100)
  question.append(str(a) + random.choice(op) + str(b))

n = random.randint(0, 99)
answer = int(input(question[n] + '=?'))
if answer == eval(question[n]):
  print('정답입니다')
else:
  print('오답입니다')
```

```python
import random

question = []
a, b = 0, 0
op = ['+', '-', '%', '*']
for i in range(100):
  a = random.randint(1, 100)
  b = random.randint(1, 100)
  o = random.randint(0, 3)
  question.append(str(a) + op[o] + str(b))

n = random.randint(0, 99)
answer = int(input(question[n] + '=?'))
if answer == eval(question[n]):
  print('정답입니다')
else:
  print('오답입니다')
```

```python
score = []
for i in range (6):
  put_score = int(input('점수를 입력하세요: '))
  score.append(put_score)

print(f'제거 전: {score}')

score.remove(max(score))
score.remove(min(score))

print(f'제거 후: {score}')
n = len(score)
sum_score = sum(score)
print(f'평균 점수: {sum_score/n}')
```
