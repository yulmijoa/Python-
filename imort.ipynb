```python
import csv

f = open('stock.csv', 'w', encoding='utf-8')
wr = csv.writer(f)
wr.writerow(['ID', '품목', '가격', '수량'])
wr.writerow(['1', '우유', '500', '10'])
wr.writerow(['2', '커피', '1000', '12'])
wr.writerow(['3', '쥬스', '2000', '5'])

f.close()
```

```python
item = input('검색을 원하는 품목을 입력하세요: ')
f= open('stock.csv', 'r')
rdr = csv.reader(f)
for line in rdr:
  if line[1] == item:
    print(line[1], '재고 수량:', line[3])
    break
f.close()
```

```python
f= open('stock.csv', 'r')
rdr = csv.reader(f)

lst=[]
for line in rdr:
  if line[1]=='우유':
    qty = int(line[3])-2
    line[3]=str(qty)
  lst.append(line)
f.close()
```

```python
f= open('stock.csv', 'w')
wr = csv.writer(f)
wr.writerows(lst)
f.close()
```
