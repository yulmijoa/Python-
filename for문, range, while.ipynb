# 시험에 이런 거 나옴
## 1~10까지 홀수 합 구하기
### range+if, range 각각

```python
sum=0
for i in range(11):
  if i%2!=0:
    sum+=i

print(sum)
```

```python
sum=0
for i in range(1, 11, 2):
  sum += i
print(sum)
```

## 1~10까지 홀수 합 구하기
### while, while+break 각각

```python
sum = 0
i = 0
while i < 11 :
  if i%2!=0:
    sum +=i
  i+= 1
print(sum)
```

```python
sum = 0
i = 1

while i < 11 :
  sum +=i
  i+= 2
print(sum)
```

```python
sum = 0
i = 1

while True :
  if i%2!=0:
    sum +=i
  i+= 2
  if i > 10:
    break
print(sum)
```

# 구구단 출력
## while문, for문

```python
x = int(input('원하는 단: '))

for i in range(1,10):
  print(f'{x}*{i}={x*i}')
```

```python
for j in range(1, 10):
  for i in range(1,10):
    print(f'{j}*{i}={j*i}')
```

```python
dan = 2
while dan<10:
  print('구구단: ', dan,'단')
  j=1
  while j<10:
    print(f'{dan}*{j}={dan*j}')
    j += 1
  dan += 1
```

```python
def max2(m, n):
  if m > n :
    return m
  else :
    return n

def min2(m, n):
  if m < n :
    return m
  else :
    return n


def avg2(m, n):
  return (m + n) / 2


num1 = int(input('1번 숫자: '))
num2 = int(input('2번 숫자: '))

print(f'{num1}과 {num2} 중에서 큰 숫자는 {max2(num1, num2)}')
print(f'{num1}과 {num2} 중에서 작은 숫자는 {min2(num1, num2)}')
print(f'{num1}과 {num2} 의 평균은 {avg2(num1, num2)}')
```
