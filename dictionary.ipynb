```python
contact = {}
contact['성']='홍'
contact['이름']='길동'
contact['직장']='율도국'

for a in contact.keys():
  print(f'{a} : {contact[a]}')
```

```python
contact = {'성': '홍', '이름':'길동', '직장':'율도국'}
for a, b in contact.items():
  print(f'{a} : {b}')
```

```python
pA = {'park', 'kim', 'lee'}
pB = {'park', 'choi'}

print(f'2개의 파티에 모두 참석한 사람: {pA.intersection(pB)}')
print(f'파티 A, B에 참석한 사람: {pA.union(pB)}')
print(f'파티 A에만 참석한 사람: {pA.difference(pB)}')
print(f'파티 A, B 중 하나에만 참석한 사람: {(pA.union(pB)).difference(pA.intersection(pB))}')
```

```python
lst = []
for i in range(1, 100):
  if i % 2 == 0 and i % 3 == 0:
    lst.append(i)
print(lst)
```

```python
lst = [i for i in range(1,100) if i%2==0 and i%3==0]
print(lst)
```

```python
oldlst = [10, 20, 30, 40, 50]
newlst = []
sum = 0
for i in oldlst:
  sum += i
  newlst.append(sum)
print(newlst)
```

```python
oldlst = [10, 20, 30, 40, 50]
newlst = []
for i in range(0, len(oldlst)):
  newlst.append(sum(oldlst[0:i+1]))
print(newlst)
```

```python
oldlst = [10, 20, 30, 40, 50]
newlst = [sum(oldlst[0:i+1]) for i in range(0,len(oldlst))]
print(newlst)
```
