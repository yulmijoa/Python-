```python
radius = 10
PI = 3.141592
area = PI * radius **2
print(f'피자면적은 {area:.1f}입니다.')
```

```python
radius2 = 20
area = PI * radius2 **2
zz = PI * 2 * radius2
print(f'피자 면적은 {area:.1f}입니다.')
print(f'피자 둘레는 {zz:.1f}입니다.')
```

```python
import ColabTurtle.Turtle as t
t.initializeTurtle()
t.shape("turtle")
n = int(input('몇각형을 그리시겠어요?(3~6): '))

for i in range(n):
  t.forward(100)
  t.left(360 //n)
```

```python
# 온도변환
f = float(input('화씨온도: '))
c = (f-32.0)*5.0/9.0
print(f"섭씨온도: {c:.1f}")
```

```python
#BMI
w = float(input('몸무게를 kg 단위로 입력하시오: '))
h = float(input('키를 cm 단위로 입력하시오: '))
bmi = (w/((h/100)**2))
print(f'당신의 bmi: {bmi:2f}')
```

```python
#자동 판매기 프로그램
a = int(input('투입한 돈: '))
b = int(input('물건 값: '))

change = a-b
print('거스름돈: ', change)
coin500s = change//500
coin100s = (change%500)//100

print("500원 동전의 개수: ", coin500s, '개')
print("100원 동전의 개수: ", coin100s, '개')
```
