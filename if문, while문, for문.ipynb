##### 주사위 게임 : 플레이어 이름 입력받고, 주사위 값은 난수를 발생시킨다. 주사위 눈 수를 비교하여 누가 더 숫자가 큰지, 아님 비겼는지 출력하라.

####while, for문
##### 같은 문제를 while이랑 for를 써서 나타내기

```python
#while문 이용
i = 1
sum = 0

while i <= 10 :
  sum += i
  i += 1

print('합계는', sum)
```

```python
#for문 이용
i = 1
sum = 0

for i in range (11):
  sum += i

print('합계는', sum)
```

```python
#짝수의 합을 더하라
i = 1
sum = 0

for i in range (11):    #or range(2, 11, 2)
  if i % 2 == 0:
    sum += i
  else:  #else + pass는 없어도 실행 됨
    pass
print('합계는', sum)
```
